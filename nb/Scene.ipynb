```python
from pathlib import Path
from ait.tools.files import imgs_from_url
from ait.tools.images import train_from_image, image_from_url
url='/home/misw/Data/AI/scenes_prod/tongue/0001'
for url_img in imgs_from_url(url):
    img = image_from_url(url_img, verbose=True)
    if img is not None:
        pil = train_from_image(img, resolutions=[512], ratios=[1.0])
        url_train = Path(url) / f'train__{url_img.name}'
        print(str(url_train))
        pil.save(url_train)
```
